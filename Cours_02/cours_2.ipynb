## REV-1 : les décorateurs 

```python
from datetime import datetime #j'importe datetime qui permet d'avoir des repère temporels. 

def chronometre(func): #je définis le décorateur qui prend en entrée une fonction, par convention func
    def wrapper(*args): #je créer une fonction wrapper qui prend un nombre indéterminé d'arguments, ceux de la fonction, et qui permet de donner des calculs à faire avant et après la fonction que l'on décore ensuite.
        debut = datetime.now() #avant de lancer la fonction, je capture le moment où la fonction est lancée 
        resultat = func(*args) #j'applique ma fonction
        fin = datetime.now() #je capture le moment de fin de la fonction
        print(f'la fonction grosse multiplication a pris : ', fin - debut) #je calcul le temps que la fonction à pris 
        return resultat #je renvoie le résultat de ma fonction
    return wrapper # je renvoie mon wrapper entier afin d'avoir les productions de ma fonction


@chronometre # je décore ma fonction
def grosse_multiplication(multiplicateur):
    multiplicande = 1000000000000000 * 9999
    resultat = 0
    for i in range(1, multiplicateur):
        resultat += (multiplicande * multiplicateur * i)
    return resultat

print(grosse_multiplication(56897 * 2)) #je lance ma fonction. 
```

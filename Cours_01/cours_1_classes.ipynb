## APP-3 et APP-4 : Création d'une classe et méthode de class

```python
class Livre :

    def __init__ (self, titre, auteur, annee_publication): #prend ces variables d'instances dont les valeurs seront à définir avec l'instance
        self.titre = titre 
        self.auteur = auteur 
        self.annee_publication = annee_publication

    def presentation (self): #création d'une méthode qui prend en compte les variables d'instances
        return(f'{self.titre} est écrit par {self.auteur} et publié en {self.annee_publication}')

LOTR = Livre("Le seigneur des Anneaux", "J.R.R Tolkien", 1954) #instanciations 
HP = Livre("Harry Potter", "J.K Rolling", 1997)


print(LOTR.presentation()) #Appel de la méthode de la class sur cette instance 
print(HP.presentation())
```

## APP-5 : Attribut de classe 

```python
class Livre : 
    nombre_total_livres = 0 # initialisation d'un compteur afin de suivre le nombre de livres créer
   
    def __init__ (self, titre, auteur, annee_publication):
        self.titre = titre 
        self.auteur = auteur 
        self.annee_publication = annee_publication

    def presentation (self): 
        Livre.nombre_total_livres += 1 #Permet d'ajouter 1 à chaque fois que cette méthode est utilisée. Comme les instances utilisent cette méthode, ça en revient à faire plus 1 quand c'est isntancié
        return(f'{self.titre} est écrit par {self.auteur} et publié en {self.annee_publication}')
        
    
    def get_nombre_total (cls) : #Ajout d'une méthode qui permet d'appeler le résultat du compteur. 
        return(f'Le nombre total de livres instanciés est {cls.nombre_total_livres}')

LOTR = Livre("Le seigneur des Anneaux", "J.R.R Tolkien", 1954)
HP = Livre("Harry Potter", "J.K Rolling", 1997)


print(LOTR.presentation())
print(HP.presentation())
print(Livre.get_nombre_total(Livre)) #Permet d'appeler la méthode get_nombre_total

# Comme get_nombre_total est un attribut de class je ne peux pas l'instancier. Pour l'appeler, j'appelle donc la classe. 
# Je fais entrer la class comme argument à ma méthode afin de pouvoir accéder à la valeur de son seul attribut de class : notre compteur.
```

## APP-6 : Comparaisons d'instances

```python
class Livre : 
    nombre_total_livres = 0 
   
    def __init__ (self, titre, auteur, annee_publication):
        self.titre = titre 
        self.auteur = auteur 
        self.annee_publication = annee_publication

    def presentation (self): 
        Livre.nombre_total_livres += 1 
        return(f'{self.titre} est écrit par {self.auteur} et publié en {self.annee_publication}')
        
    
    def get_nombre_total (cls) :  
        return(f'Le nombre total de livres instanciés est {cls.nombre_total_livres}')

    def __eq__(self, autre): #self défini l'objet se situant à gauche des signes ==, et other l'objet à droite. On pourrait les renomer autrements mais je suis la "convention"
        if not isinstance(autre, Livre): #si l'objet de droite n'est une instance de Livre
            return False #permet ainsi de signifier que les objets donnés en entrées ne sont pas comparable. 
        
        return (self.titre == autre.titre and   
                self.auteur == autre.auteur and
                self.annee_publication == autre.annee_publication
                ) #Si les objets sont comparables, on les compares points par points. Si ils sont identiques, il retourne True. Sinon, ce sera False


livre1 = Livre("Le seigneur des Anneaux", "J.R.R Tolkien", 1954)
livre2 = Livre("Harry Potter", "J.K Rolling", 1997)
livre3 = Livre("Le seigneur des Anneaux", "J.R.R Tolkien", 1954)


print(livre1.presentation())
print(livre2.presentation())
print(Livre.get_nombre_total(Livre)) 
print(livre1 == livre2) # cette syntaxe permet d'appeler __eq__ dans l'instance. On pourrait aussi écrire print(livre1.__eq__(livre2)) mais c'est moins synthétique bien que plus transparent.
print(livre1==livre3) # imprimera True, ce sont exactements les mêmes valeurs d'attributs d'instances. 
```
